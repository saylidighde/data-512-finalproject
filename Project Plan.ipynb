## Introduction

In the competitive global economy today, science, technology, entrepreneurship, and innovation play an important role in the sustainable development of any nation. During my undergraduate education in Information Technology, I was involved with an NGO, Foundation for Ecological Security (FES) for sustainable development of rural Gujarat. I want to understand the effects of Research and development expenditure as % of GDP on the development of science, technology, and business on a country. I plan to analyze the data for last 15 to 20 years. As I have a strong interest in research and information science and also aspire to be an entrepreneur, I am curious to know if deliberate steps by the government positively affect scientific research, entrepreneurship, and number of patents rather than growth being a random phenomenon. My goal is to address the human-centered research questions listed in the Analysis section. Later, I might dabble in the environment and sustainable development datasets to see if research expenditure positively affects growth in these areas too.

Hence, I chose the World Development Indicators (States and Markets) dataset by the World Bank for my exploratory analysis.


## General Plan

According to [World Bank](http://datatopics.worldbank.org/world-development-indicators/), The World Development Indicators is "a compilation of relevant, high-quality, and internationally comparable statistics about global development and the fight against poverty. The database contains 1,600 time series indicators for 217 economies and more than 40 country groups, with data for many indicators going back more than 50 years."

It has 6 data themes:
1. Poverty and Inequality
2. People 
3. Environment
4. Economy
5. States and Markets
6. Global Links

Out of these, I will be mainly focusing on the theme "States and Markets" for indicators like Research and development expenditure (% of GDP), Scientific and technical journal articles, ICT goods exports (% of total goods exports), Patent applications (residents), and Time required to start a business (days).

#### States and Markets
According to World Bank, States and markets statistics focus on "private sector development, including entrepreneurship and finance; infrastructure development; and public policies and institutions that affect the economy. Many of these topics are also part of the Sustainable Development Goals."

Again, there are datasets for various economic indicators. Currently, I have downloaded the datasets for the following indicators:

1. Time required to start a business (in days)

    Number of calendar days to complete the procedures to legally operate a business. If a procedure can be speeded up at additional cost, the fastest procedure, independent of cost is chosen.

    This indicator has 3 CSV files. Schema of each of the files is as follows:
       
    * WDIs
    Country Name
    Country Code
    Indicator Name
    Indicator Code
    Column-wise indicator values from 1960-2017

    * Country metadata
    Country Code
    Region
    Income Group
    Special Notes
    Table Name

    * Indicator metadata
    Indicator Code
    Indicator Name
    Source Note
    Source Organization
    
2. Research and development expenditure (% of GDP)

    Gloss domestic expenditures on research and development, expressed as a percent of GDP. They include both capital and current expenditures in the 4 main sectors: business enterprise, government, higher education, and private non-profit. It covers basic and applied research as well as experimental development.

    Again, this indicator has 3 CSV files. Schema of each of the files is as follows:   

    * WDIs
    Country Name
    Country Code
    Indicator Name
    Indicator Code
    Column-wise indicator values from 1960-2017

    * Country metadata
    Country Code
    Region
    Income Group
    Special Notes
    Table Name

    * Indicator metadata
    Indicator Code
    Indicator Name
    Source Note
    Source Organization

Similarly, other indicator datasets like Scientific and technical journal articles, ICT goods exports (% of total goods exports), and Patent applications (residents) followed the same format as above listing out the indicator values country-wise from 1960 till present.


## Analysis Plan

#### Probable research questions:

1. Does Research and development expenditure (% of GDP) affect the number of Patent applications by residents in a country? 
2. Does Research and development expenditure (% of GDP) affect the ICT goods exports (% of total goods exports) in a country?
3. Is there a correlation between Research and development expenditure (% of GDP) and Scientific and technical journal articles published?
4. How do Industrial design applications vary with Research and development expenditure (% of GDP) and the time required to start a business? 
5. How has Research and development expenditure (% of GDP) changed over the last 15 years?
6. 10 countries with the highest and lowest Research and development expenditure (% of GDP) in last 15 years
7. Which countries are efficient with Research and development expenditure (% of GDP) by getting more journal articles published in relatively less expenditure?

#### Corresponding hypotheses:

1. Yes, R&D expenditure positively affects the number of Patent applications.
2. Yes, R&D expenditure positively affects the the ICT goods exports.
3. Yes, there is a positive correlation between R&D expenditure and the number of Scientific and technical journal articles published.
4. It is more of an exploratory question. 
5. R&D expenditure has increased with time
6. N/A - Exploratory question. 
7. N/A - Exploratory question. 

*In all the above hypotheses, R&D expenditure means R&D expenditure as % of GDP.

I don't really foresee any unknowns or dependencies that might the completion of this project. I might have to deal with missing values while joining the above datasets. Also, as the GDP of different countries can be vastly different, I may need to consider actual R&D expenditure as part of GDP instead of just R&D expenditure as % of GDP. 

The reaserch questions are more exploratory in nature and I am curious if I find any interesting insights in the data during exploratory data analysis. Like, are the advanced nations engaged in systematic efforts for technological advancement? Are countries showing positive correlation between ease of setting up businesses and the R&D indicator more inclined towards technological businesses? Does Information and Communication goods export depend on the research and development of a country? Or are some nations ranking high in this area while simultaneously ranking low in overall R&D and patent filing? Are they just providing cheap information technology labor? I also expect to find and explain interesting outliers in the analysis where scientific advancement eludes any usual government support schemes. 

Although I don't expect these findings to result in any groundbreaking research, I hope this exploratory project enhances my understanding of the effects of the R&D expenditure indicator on the development of science, research, technology exports, and entrepreneurship in a country.  

I will be extensively using the Pandas library from Python for data munging and cleaning. For the visualizations, I will be using the Seaborn and Matplotlib libraries. I will be joining the dataframes obtained from the various CSV files for each of the indicators to each other for my analysis. I will be using a Jupyter notebook for documentation and make the reasearch reporoducible. This would allow the scientific as well as the non-scientific communities to reuse and build on my research.
